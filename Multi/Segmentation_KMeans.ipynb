```python
# import required libraries
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from skimage import io
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# load and read image
image = io.imread(r"C:\Users\Arfan Shah\potato_data\multi\Train\Potato___Late_blight\1a68ab56-2000-4d7a-ad4a-88d300104b7d___RS_LB 2594.JPG")
plt.imshow(image)
```

```python
# shape of image
image.shape
```

```python
reshaped_image = image.reshape(-1, 3)
```

```python
# apply kmeans clustering
kmeans = KMeans(n_clusters = 15)

# fit image array
kmeans.fit(reshaped_image)
```

```python
# cluster centers
kmeans.cluster_centers_
```

```python
# find labels
labels = kmeans.labels_
```

```python
# now apply segmentation
segmented_image = kmeans.cluster_centers_[labels]
segmented_image
```

```python
# normalize data
segmented_image = segmented_image/255.0
```

```python
# convert into original image
segmented_image = segmented_image.reshape(image.shape)
```

```python
# segmented image
plt.imshow(segmented_image)
```
