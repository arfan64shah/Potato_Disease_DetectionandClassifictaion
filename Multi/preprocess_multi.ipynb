```python
import matplotlib.pyplot as plt
from skimage import io, color, filters, morphology, segmentation

# Load the leaf image
image_path = r"C:\Users\Arfan Shah\potato_data\multi\Train\Potato___Early_blight\0e0a1b51-f61c-4934-bc57-a820af1faacb___RS_Early.B 7147.JPG"
leaf_image = io.imread(image_path)

# Convert the image to grayscale
gray_image = color.rgb2gray(leaf_image)

# Apply thresholding to segment the affected parts
threshold_value = filters.threshold_otsu(gray_image)
binary_image = gray_image > threshold_value

# Clean up the binary image using morphological operations
cleaned_image = morphology.remove_small_objects(binary_image, min_size=1000)

# Label connected components in the cleaned image
labeled_image = segmentation.label(cleaned_image)

# Plot the original image, grayscale image, and the affected parts
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(leaf_image)
axes[0].set_title('Original Image')

axes[1].imshow(gray_image, cmap='gray')
axes[1].set_title('Grayscale Image')

axes[2].imshow(labeled_image, cmap='nipy_spectral')
axes[2].set_title('Affected Parts')

for ax in axes:
    ax.axis('off')

plt.show()
```
